###Vamos criar uma dataframe novo para usar como exemplo de como podemos preencher esses dados nulos

Criamos o nova df.

- `.fillna()` = preencher  
- `.mean( )` = média  
- `.round(2)` = arredondando com 2 casas decimais

Preenchemos os dados nulos com a média dos salarios e com a mediana, criando uma coluna nova para cada coisa.

````python
import numpy as np

# Criação de um dataframe de exemplo
df_salarios = pd.DataFrame({
    'nome': ['João', 'Maria', 'Pedro', 'Ana', 'Enzo'],
    'idade': [25, 30, 35, 40, 19],
    'salario': [4000, np.nan, 5000, np.nan, 100000]
})

# Preenchendo os dados nulos com a media dos salários, criando uma nova coluna
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# Preenchendo com a mediana dos salarios, criando mais uma coluna
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
````

Criamos um novo dataframe para fazer um novo exemplo usando temperaturas.

Dessa vez criamos um nova tabela preenchendo os dados nulos com dados anteriores a eles. `.ffill()`

Os dados nulos foram preenchidos com 30 pois o dados anterior a eles é 30.

````python
import numpy as np

df_temperaturas = pd.DataFrame({
    'Dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sab', 'Dom'],
    'Temperatura': [30, np.nan, np.nan, 31, 27, np.nan, 20]
})

# Dessa vez criamos um nova tabela preenchendo os dados nulos com dados anteriores a eles.
df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()

df_temperaturas
````

`.bfill()` é o contratio do `.ffill()`.

Ele vai preencher com o dados seguindo do dado nulo.

````python
import numpy as np

df_temperaturas = pd.DataFrame({
    'Dia': ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sab', 'Dom'],
    'Temperatura': [30, np.nan, np.nan, 31, 27, np.nan, 20]
})

# Dessa vez criamos um nova tabela preenchendo os dados nulos com dados anteriores a eles.
df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()

df_temperaturas
````

- Crie um gráfico de média salarial por senioridade em barras usando o plotly


````python
import plotly.express as px
# Calcular a média salarial por nível de experiência
media_salario_por_nivel = df_limpo.groupby('nivel_experiencia')['salario_em_usd'].mean().reset_index()

# Ordenar os níveis de experiência
ordem_nivel_experiencia = ['Junior', 'Pleno', 'Senior', 'Executivo']
media_salario_por_nivel['nivel_experiencia'] = pd.Categorical(media_salario_por_nivel['nivel_experiencia'], categories=ordem_nivel_experiencia, ordered=True)
media_salario_por_nivel = media_salario_por_nivel.sort_values('nivel_experiencia')


# Criar o gráfico de barras interativo
fig = px.bar(media_salario_por_nivel,
             x='nivel_experiencia',
             y='salario_em_usd',
             title='Média de Salário por Nível de Experiência',
             labels={'nivel_experiencia': 'Nível de Experiência', 'salario_em_usd': 'Salário médio anual (USD)'})

fig.show()
````

#### Outra forma de fazer

Precisamos importar essa biblioteca para converter siglas de 2 letras para 3 letras

Vamos fazer a conversão


````python
import pycountry
# Função para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia_empregado'].apply(iso2_to_iso3)
````

Agora o gráfico

````python
# Calcular média salarial por país (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['salario_em_usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='salario_em_usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'salario_em_usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
````
